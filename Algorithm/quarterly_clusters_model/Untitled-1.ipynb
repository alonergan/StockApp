```python
import pandas as pd

ratios_df = pd.read_csv('all_stock_ratios.csv')

# for testing, only use some stock(s)
ratios_df = ratios_df[ratios_df['Ticker'].isin(['UE','HALO', 'ACHC', 'AEE', 'HCKT', 'MA'])]
# ratios_df = ratios_df[ratios_df['Ticker'].isin(['AEE'])]
```

```python
# show the ratios
# plotting individual ratios over time for each stock
import matplotlib.pyplot as plt



for ratio in ratios_df.columns[2:]:
    plt.figure(figsize=(10, 6))
    for stock in ratios_df['Ticker'].unique():
        stock_data = ratios_df[ratios_df['Ticker'] == stock]
        plt.plot(stock_data['fiscalDateEnding'], stock_data[ratio], label=stock)
    plt.title(f'{ratio} over time')
    plt.xlabel('Date')
    plt.ylabel(ratio)
    plt.legend()
    plt.show()
```
