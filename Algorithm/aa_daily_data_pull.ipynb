```python
# read api key from file
with open("api_key.txt", "r") as f:
    api_key = f.read().strip()
```

```python
# get list of stocks from file
with open("ticker_list.csv", "r") as f:
    stocks = [line.strip() for line in f.readlines()]
    # ignore header
    stocks = stocks[1:]
```

```python
import requests

def api_call(function, symbol):
    url = f'https://www.alphavantage.co/query?function={function}&symbol={symbol}&apikey={api_key}&datatype=csv&outputsize=full'
    # url = f'https://www.alphavantage.co/query?function={function}&symbol={symbol}&apikey={api_key}&datatype=csv'
    r = requests.get(url)
    data = r.text
    # add ticker symbol to each line of data that is not the header
    data = data.splitlines()
    header = data[0]
    data = [f'{symbol},{line}' for line in data]

    data[0] = f'ticker,{header}'

    data = '\n'.join(data)

    return data
```

```python
print(api_call('TIME_SERIES_DAILY', 'AAPL'))
```

```python
# for testing, only get data for some stocks
# stocks = stocks[:90]

# takes about 4 minutes to run with all stocks.

import time # need a timer due to API call limits, 75 a minute

# with open('aa_daily_data.csv', 'w+') as f:
with open('aa_daily_data_2026-04-07.csv', 'w+') as f:
    for stock in stocks:
        f.write(api_call('TIME_SERIES_DAILY', stock)+ '\n')
        time.sleep(0.81) # sleep for 0.81 seconds to avoid hitting API call limit, they do not like burst of calls so this
                         # will space them out.
```
