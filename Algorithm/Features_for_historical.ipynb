```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df = pd.read_excel("combined_data.csv")
```

```python
df['Date'] =pd.to_datetime(df['Date'])
df = df.sort_values(['Ticker','Date'])
```

```python
df['MA_5'] =df.groupby('Ticker')['Close'].transform(
    lambda x: x.rolling(window=5, min_periods=1).mean()
)
```

```python
print(df[['Date', 'Ticker','Close','MA_5']].head(10))
```

```python
def create_features(df):
  group = df.copy()
  group['MA5'] =group.groupby('Ticker')['Close'].transform(
      lambda x: x.rolling(window=5, min_periods=1).mean()
  )

  group['Daily_Return' ] = group.groupby('Ticker')['Close'].transform(
      lambda x: x.pct_change())

  return group
```

```python
df_with_features = df.groupby('Ticker')[df.columns].apply(create_features)
df_with_features.dropna(inplace=True)
df_with_features.reset_index(drop=True, inplace=True)
df_with_features.head()
```
