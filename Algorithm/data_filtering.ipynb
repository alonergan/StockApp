```python
import yfinance as yf
import pandas as pd

def get_quarterly(ticker):
    try:
        if not ticker or not isinstance(ticker, str):
            print(f"Invalid ticker: {ticker}")
            return None
        ticker_obj = yf.Ticker(ticker)
        income = ticker_obj.quarterly_financials
        balance_sheet = ticker_obj.quarterly_balance_sheet
        cash_flow = ticker_obj.quarterly_cashflow
        earnings = ticker_obj.quarterly_income_stmt

        if income is None or balance_sheet is None or cash_flow is None or earnings is None:
            print(f"No data for {ticker}")
            return None
        return {
            "income": income,
            "balance_sheet": balance_sheet,
            "cash_flow": cash_flow,
            "earnings": earnings
        }
    except Exception as e:
        print(f"Error with {ticker}: {e}")
        return None



```

```python
Quarterly_data=[]

for ticker in tickers['Ticker']:
    print("processing ticker: ",ticker)
    data = get_quarterly(ticker)

    if data is None:
        print(f"No data for {ticker}")
        continue

    income = data["income"].T
    balance = data["balance_sheet"].T
    cash = data["cash_flow"].T


    combined = pd.concat([income, balance, cash], axis=1)


    combined["Ticker"] = ticker

    Quarterly_data.append(combined)


final_dataset = pd.concat(Quarterly_data)


final_dataset = final_dataset.reset_index().rename(columns={"index": "Quarter"})

print(final_dataset.head())
```

```python
final_dataset.to_csv("quarterly_financials.csv", index=False)
```

```python
#total liabilities/net worth
#working capital/sales
#pretax net nonop .inc/sales
#(net  inc-Divid)/Net inc
#(net inc -inc taxes)/netinc
#sales/total assets
#inc. common sh./net worth
#cash flow/num shares
#inventory/tot curr liab
#earnings/price
#relative price evaluation



1. Revenue / Sales

2. Net Income

3. Pretax Non-Operating Income

4. Income Taxes

5. Dividends Paid

6. Income Available to Common Shareholders

7. Total Assets

8. Total Liabilities

9. Shareholders’ Equity / Net Worth

10. Current Assets

11. Current Liabilities

12. Inventory

13. Operating Cash Flow

14. Shares Outstanding

15. Stock Price

Historical Stock Price Series

Total Liabilities / Net Worth = Total Liabilities ÷ Shareholders’ Equity

Working Capital / Sales = (Current Assets - Current Liabilities) ÷ Revenue

Pretax Net Non-Operating Income / Sales = Pretax Non-Operating Income ÷ Revenue

(Net Income -Dividends) / Net Income = (Net Income - Dividends Paid) ÷ Net Income

(Net Income - Income Taxes) / Net Income = (Net Income - Income Taxes) ÷ Net Income

Sales / Total Assets = Revenue ÷ Total Assets

Income to Common Shareholders / Net Worth = Income Available to Common Shareholders ÷ Shareholders’ Equity

Cash Flow / Number of Shares = Operating Cash Flow ÷ Shares Outstanding

Inventory / Total Current Liabilities = Inventory ÷ Current Liabilities

Earnings / Price = (Net Income ÷ Shares Outstanding) ÷ Stock Price

Relative Price Evaluation = Stock Price ÷ Historical Stock Price Benchmark (e.g., average or index price)
```

```python
import yfinance as yf
import pandas as pd

def get_company_data(ticker):

    t = yf.Ticker(ticker)

    income = t.quarterly_financials.T
    balance = t.quarterly_balance_sheet.T
    cashflow = t.quarterly_cashflow.T

    df = pd.concat([income, balance, cashflow], axis=1)

    columns_needed = [
    "Quarter",
    "Ticker",
    "Total Revenue",
    "Net Income",
    "Pretax Income",
    "Tax Provision",
    "Cash Dividends Paid",
    "Net Income Common Stockholders",
    "Total Assets",
    "Total Liabilities Net Minority Interest",
    "Stockholders Equity",
    "Current Assets",
    "Current Liabilities",
    "Inventory",
    "Operating Cash Flow",
    "Shares Outstanding",
    "Stock Price"
    ]

    df = df[[c for c in columns_needed if c in df.columns]]

    df["Ticker"] = ticker
    hist = t.history(period="max")
    quarter_prices = hist["Close"].resample("Q").last()
    quarter_prices.index = quarter_prices.index.to_period("Q")

    df.index = df.index.to_period("Q")

    df["Stock Price"] = df.index.map(quarter_prices)

    return df


all_data = []

for ticker in tickers["Ticker"]:
    try:
        data = get_company_data(ticker)
        all_data.append(data)
    except Exception as e:
        print(ticker, e)

financial_dataset = pd.concat(all_data)

financial_dataset = financial_dataset.reset_index().rename(columns={"index":"Quarter"})
```

```python
#fix missing in invintory column

import yfinance as yf
import pandas as pd
import numpy as np
from datetime import datetime
import time

def get_inventory_from_yahoo(ticker, quarter_date):
    """
    Get inventory value for a specific ticker and quarter from Yahoo Finance
    """
    try:

        stock = yf.Ticker(ticker)

        quarterly_financials = stock.quarterly_balance_sheet

        if quarterly_financials is not None and not quarterly_financials.empty:

            inventory_names = ['Inventory', 'Total Inventory', 'Inventories',
                              'Net Inventory', 'Inventory Net','Current Assets']


            inventory_row = None
            for name in inventory_names:
                if name in quarterly_financials.index:
                    inventory_row = quarterly_financials.loc[name]
                    break

            if inventory_row is not None:
                if isinstance(quarter_date, str):
                    year = int(quarter_date[:4])
                    quarter = int(quarter_date[5:])
                    month = (quarter - 1) * 3 + 1
                    date = datetime(year, month, 1)
                else:
                    date = quarter_date


                if date in inventory_row.index:
                    return inventory_row[date]
                else:
                    dates = inventory_row.index
                    if len(dates) > 0:
                        closest_date = min(dates, key=lambda x: abs((x - date).days))
                        return inventory_row[closest_date]

        return None

    except Exception as e:
        print(f"Error fetching data for {ticker} on {quarter_date}: {e}")
        return None
```

```python
import yfinance as yf
import pandas as pd

def get_daily(ticker):
    t = yf.Ticker(ticker)
    data = t.history(period="5d")

    if data.empty:
        return None


    data = data.reset_index()
    prev_day = data.iloc[-2]


    prev_day_info = {
        "Ticker": ticker,
        "Date": prev_day["Date"],
        "Open": prev_day["Open"],
        "High": prev_day["High"],
        "Low": prev_day["Low"],
        "Close": prev_day["Close"],
        "Volume": prev_day["Volume"]
    }

    return prev_day_info


if __name__ == "__main__":
    df = pd.read_csv("combined_data.csv")
    results = []

    tickers = pd.DataFrame(df['Ticker'].unique(), columns=['Ticker'])

    for ticker in tickers['Ticker']:
          daily = get_daily(ticker)
      if daily:
          results.append(daily)

    daily_df = pd.DataFrame(results)

```
